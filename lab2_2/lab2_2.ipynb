```python
import numpy as np
import matplotlib.pyplot as plt

data_1 = np.loadtxt('lab4_1.txt')
data_2 = np.loadtxt('lab4_2.txt')
data_3 = np.loadtxt('lab_5.txt')
data_4 = np.loadtxt('lab_6.txt')
```

# Lab_1

```python
U_p, U_n = data_1[:, 0], data_1[:, 1]
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel(r'$U_п, В$')
plt.ylabel(r'$U_н, В$')
plt.plot(U_p, U_n)
plt.savefig('graphs/Амплитудная_характеристика_1.png', dpi=300)
```

## Lab_4_1 $K_{ст}$

```python
K_st = ((np.max(U_p) - np.min(U_p)) / np.mean(U_p)) / ((np.max(U_n) - np.min(U_n)) / np.mean(U_n))
print(f"{K_st=}")
```

# Lab_2

```python
R_n, U_n, I_n, I_st = data_2[:, 0], data_2[:, 1], data_2[:, 2], data_2[:, 3]
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel(r'$I_н, мА$')
plt.ylabel(r'$U_н, В$')
plt.plot(I_n, U_n)
plt.savefig('graphs/Выходная_характеристика_2.png', dpi=300)
```

## Lab_4_2 Подсчёт кпд, $P_{вх}$ и $P_{н}$

```python
P_vh = (I_n + I_st) * 9.21
P_n = I_n * U_n
print("R_n\tU_n\tI_n\tI_st\tP_vh\tP_n\tкпд")
for i in range(8):
    print(f"{round(float(R_n[i]), 2)}\t{round(float(U_n[i]), 2)}\t{round(float(I_n[i]), 2)}\t{round(float(I_st[i]), 2)}\t{round(float(P_vh[i]), 2)}\t{round(float(P_n[i]), 2)}\t{round(float(P_n[i]) / float(P_vh[i]), 2)}")
```

## Lab_4_2 Расчёт $R_{вых}$

```python
R_exit = ((np.max(U_n) - np.min(U_n)) / (np.max(I_n) - np.min(I_n))) * 1000
print(f"{R_exit=}")
```
