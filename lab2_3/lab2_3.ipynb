```python
import numpy as np
import matplotlib.pyplot as plt

data_1 = np.loadtxt('1.txt')
data_2 = np.loadtxt('2.txt')
data_3 = np.loadtxt('3.txt')
data_4 = np.loadtxt('4.txt')
data_5 = np.loadtxt('5.txt')
```

# Расчёт $K_{п}$ для т. 7.1
## $K_{п} = \frac {U_{пульс}} {U_н}$

```python
C_1, U_2, U_n, I_n, U_p = data_1[:, 0], data_1[:, 1], data_1[:, 2], data_1[:, 3], data_1[:, 4]
print("C_1\tU_2\tU_n\tI_н\tU_п\tK_п")
for i in range(8):
    print(f"{C_1[i]}\t{U_2[i]}\t{U_n[i]}\t{I_n[i]}\t{U_p[i]}\t{np.round(U_p[i] / U_n[i], 3)}")
```

# т. 7.2: построение $U_н=f(I_н)$, $K_{п} = \frac {U_{пульс}} {U_н}$.

```python
R_n, U_2, U_n, I_n, U_p = data_2[:, 0], data_2[:, 1], data_2[:, 2], data_2[:, 3], data_2[:, 4]
print("R_н\tU_2\tU_n\tI_н\tU_п\tK_п")
for i in range(8):
    print(f"{R_n[i]}\t{U_2[i]}\t{U_n[i]}\t{I_n[i]}\t{U_p[i]}\t{np.round(U_p[i] / U_n[i], 3)}")
```

```python
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel(r'$I_н, мА$')
plt.ylabel(r'$U_н, В$')
plt.plot(I_n, U_n)
plt.savefig('graphs/Выходная_характеристика_7_2.png', dpi=300)
```
