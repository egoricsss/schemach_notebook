## Амплитудная характеристика для инфертирующей схемы при постоянном токе
- Максимальное значение $e_{c}$ составило $10.03 B$ 
- $U_{вых}=f(e_{c}), I=const$

```python
e_c, U_v = data_1[:, 0], data_1[:, 1]
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel('e_c [В]')
plt.ylabel('U_вых [В]')
plt.plot(e_c, U_v)
plt.savefig('частотная_характеристика_постоянный_1.png')
```

## Амплитудная характеристика для инвертирующей схемы при переменном токе
- Максимальное значение $e_{c}$ составило $9.02 B$

```python
e_c, U_v = data_2[:, 0], data_2[:, 1]
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel('e_c [В]')
plt.ylabel('U_вых [В]')
plt.plot(e_c, U_v)
plt.savefig('частотная_характеристика_переменный_1.png')
```

## Теоритическая и эксперементальная зависимости коэффициента усилинения
- Формула для теоритической зависимости: $K = -\frac{R_{2}}{R_{1}}, R_{1} = const$ 
- Формула для экспериментальной зависимости $K = \frac{U_{вых}}{U_{вх}}$

```python
e_c, U_v, R_2 = data_3[:, 0], data_3[:, 1], data_3[:, 2]
plt.figure(figsize=(10, 6))
plt.plot(R_2, R_2 / 10, label='теоритическая зависимость')
plt.plot(R_2, U_v / e_c, label='экспериментальная зависимость')
plt.grid()
plt.legend()
plt.ylabel('K')
plt.xlabel('R_2 [кОм]')
plt.savefig('K.png')
```
