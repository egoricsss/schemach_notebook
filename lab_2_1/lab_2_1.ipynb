# Зависимости + считывание данных

```python
import numpy as np
import matplotlib.pyplot as plt

data_1 = np.loadtxt('VD5.txt')
data_2 = np.loadtxt('VD6.txt')
data_3 = np.loadtxt('VD7.txt')
data_4 = np.loadtxt('HL1.txt')
```

# ВАХ VD5

```python
U, I = data_1[:, 0], data_1[:, 1]
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel('U_VD5 [В]')
plt.ylabel('I_VD5 [мА]')
plt.plot(U, I)
plt.savefig('graphs/ВАХ_VD5.png', dpi=300)
```
